```python
import albumentations as A 
import numpy as np
import cv2
```

```python
xa_transform =  A.Compose([
                A.ShiftScaleRotate(
                    shift_limit=0.15, scale_limit=0.15, rotate_limit= 0.10 , p=0.5),
                A.RGBShift(r_shift_limit= 8, g_shift_limit=8, b_shift_limit=8, p=0.5),
                A.RandomBrightnessContrast(p=0.2),
                A.MultiplicativeNoise(multiplier=1.5, p=0.5),



                A.RGBShift(r_shift_limit=8, g_shift_limit=8, b_shift_limit=8, p=0.3),
                A.ShiftScaleRotate(
                    shift_limit=0.3, scale_limit=0.15, rotate_limit= 0.10 , p=0.5),
                    
                A.MultiplicativeNoise(multiplier=1, p=0.5),

                # A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.216)),
                # A.Resize(height=image_shape[0], width=image_shape[1]),
                
                A.RandomRotate90(p = 0.5),
                
                A.MotionBlur(7, p = 0.4 ),
                A.ShiftScaleRotate(
                    shift_limit=0.15, scale_limit=0.1, rotate_limit= 0.10 , p=0.5),

                # add noise by albummentations
                A.MultiplicativeNoise(multiplier=1.5, p=0.5),
                A.RandomBrightnessContrast(p=0.2),
                A.RandomGamma(p=0.2),
                # ShiftScaleRotate by albumentations
                A.ShiftScaleRotate(shift_limit=0.15, scale_limit=0.15, rotate_limit= 0.10 , p=0.5),
                A.MultiplicativeNoise(multiplier=1, p=0.5)
            
            
            ])
```

```python
pill_path = "/mnt/disk1/doan/xanh/pill_big_xa/data_pill/VAIPE_P_1164_14107.jpg"
img_pill = cv2.imread(pill_path)
```

```python
import matplotlib.pyplot as plt
for i in range(20):
    # apply transform to image from albumentations library
    img_new = xa_transform(image=img_pill)['image']
    # convert RGB to BGR
    img_new = cv2.cvtColor(img_new, cv2.COLOR_RGB2BGR)
    plt.imshow(img_new)
    plt.show()

```
